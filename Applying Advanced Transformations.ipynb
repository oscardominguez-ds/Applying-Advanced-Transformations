```python
# imports
import pandas as pd

# Numpy
import numpy as np
## Importing the OS and JSON Modules
import os,json

# Open saved file and preview again
sh_info_df = pd.read_csv("Data/superhero_info - superhero_info.csv", low_memory = False)
sh_info_df.head()
```

# First dataframe 

```python
#Info, no missing values
sh_info_df.info()
```

## Duplicates

```python
#Printing duplicates
print(f'There are {sh_info_df.duplicated().sum()} duplicate rows.')
```

```python
# Example 4: Split single column into two columns use apply()
# Resource - https://sparkbyexamples.com/pandas/pandas-split-column/
sh_info_df[['Hero', 'Publisher']] \
= sh_info_df["Hero|Publisher"].apply(lambda x: pd.Series(str(x).split("|")))      

sh_info_df.drop("Hero|Publisher", axis = 1, inplace = True)

sh_info_df
```

```python
# create 2 new columns, FirstName and LastName by splitting the Name column
sh_info_df[['height cm','weight kg']] = sh_info_df['Measurements'].str.split(',', expand=True)
# drop the 'Name' column
sh_info_df.drop('Measurements', axis=1, inplace=True)
sh_info_df.head()
```

```python
# Resource - 
# https://stackoverflow.com/questions/13682044/remove-unwanted-parts-from-strings-in-a-column

sh_info_df["height cm"] = sh_info_df["height cm"].\
    map(lambda x: x.lstrip("{'Height': '").rstrip("cm'"))
sh_info_df.head()
```

```python
# Resource -
# https://stackoverflow.com/questions/13682044/remove-unwanted-parts-from-strings-in-a-column
sh_info_df["weight kg"] = sh_info_df["weight kg"].\
    map(lambda x: x.lstrip("'Weight': '").rstrip("kg'}"))
sh_info_df.head()
```

```python
#Converting 'height cm' column to float
sh_info_df = sh_info_df.astype({'height cm':'float'})
sh_info_df.dtypes
```

```python
#Converting 'weight kg' column to float
sh_info_df = sh_info_df.astype({'weight kg':'float'})
sh_info_df.dtypes
```

# Second dataframe

```python
#Creating new column 'Powers_split' to save list of powers
# sh_powers_df['Powers_split']= sh_powers_df['Powers'].str.split(",",expand=False)
# sh_powers_df.head(2)

sh_powers_df['Powers_split']= sh_powers_df['Powers'].str.split(",",expand=False)
sh_powers_df.head(2)
```

```python
# exploding 'Powers_split' to have a column for each power
exploded = sh_powers_df.explode('Powers_split')
exploded.head(5)
```

```python
## saving the unique values from the exploded column
cols_to_make = exploded['Powers_split'].dropna().unique()
cols_to_make
```

```python
# Creating columns for each super power
for col in cols_to_make:
    sh_powers_df[col] = sh_powers_df['Powers'].str.contains(col)
sh_powers_df.head()
```

```python
# dropping columns that were used to one hot encode
sh_powers_df = sh_powers_df.drop(columns=['Powers', 'Powers_split'])
```

```python
# I was getting null values when I first concatenated. I searched Google and found that
# I needed to include axis = 1
# Resource - https://intellipaat.com/community/32857/pandas-concat-generates-nan-values

df = pd.merge(sh_info_df, sh_powers_df, left_on='Hero', right_on ='hero_names')

df.head()
```

```python
# I noticed some columns were not showing that were in the second dataframe so I printed 
# all of them
#Resource to iterate through columns
# https://www.geeksforgeeks.org/how-to-get-column-names-in-pandas-dataframe/
    
# iterating the columns
for col in df.columns:
    print(col)
```

```python
#Saved concatenated files
df.to_csv('Data/superhero_data.csv.gz',compression='gzip',index=False)
```

# Compare the average weight of super powers who have Super Speed to those who do not.


```python
#Using groupby to organize by True or False using column Super Speed
avg_weight = df.groupby('Super Speed')['weight kg'].mean()
avg_weight
```

# What is the average height of heroes for each publisher?

```python
#Using groupby to sory by Publisher name the average height
avg_height = df.groupby('Publisher')['height cm'].mean()
avg_height
```
